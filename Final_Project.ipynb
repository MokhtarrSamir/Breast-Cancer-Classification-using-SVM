```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel("breast-cancer-wisconsin.xlsx")
```

## Handling Missing Value

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop(columns=['ID'], inplace=True)
```

```python
data.head()
```

### Encoding

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['Diagnosis'] = encoder.fit_transform(data['Diagnosis'])
data['Diagnosis'].value_counts()
```

### Shuffling & Splitting

```python
from sklearn.model_selection import train_test_split

x = data.drop('Diagnosis', axis=1)
y = data['Diagnosis']

x_train, x_temp, y_train, y_temp = train_test_split(x, y, train_size=400, shuffle=True, random_state=42)
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=69, shuffle=True)
```

### Preprocessing

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train[x_train.columns] = scaler.fit_transform(x_train)
x_val[x_val.columns] = scaler.transform(x_val)
x_test[x_test.columns] = scaler.transform(x_test)
```

### Feature Selection

```python
# Compute correlation of each feature with the target
correlations = data.corr()['Diagnosis'].drop('Diagnosis').abs().sort_values(ascending=False)

# Select top N features (for example, top 10)
top_features = correlations.index.tolist()[:2]
print("Top correlated features:", top_features)
```

```python
# Select features
x_train = x_train[top_features]
x_val = x_val[top_features]
x_test = x_test[top_features]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

feat1, feat2 = top_features
sns.scatterplot(data=x_train, x=feat1, y=feat2, hue=y_train)
plt.show()
```

## Classification using Logistic Regression & SVM

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
lr_model = LogisticRegression()
svm = SVC(kernel='linear')

lr_model.fit(x_train, y_train)
svm.fit(x_train, y_train)

y_pred = lr_model.predict(x_test)
print("LogisticRegression:", accuracy_score(y_test, y_pred))
y_pred = svm.predict(x_test)
print("SVM:", accuracy_score(y_test, y_pred))
```

## Kernels & Decision Boundaries

```python
from svm_margin_plot import plot_svm_boundary
```

```python
model_linear = SVC(kernel='linear')
model_linear.fit(x_train, y_train)

plot_svm_boundary(model_linear, x_train, y_train)
```

```python
model_poly = SVC(kernel='poly')
model_poly.fit(x_train, y_train)

plot_svm_boundary(model_poly, x_train, y_train)
```

```python
model_rbf = SVC(kernel='rbf')
model_rbf.fit(x_train, y_train)

plot_svm_boundary(model_rbf, x_train, y_train)
```

```python
y_val_pred = model_linear.predict(x_val)
y_test_pred = model_linear.predict(x_test)
print('Linear Accuracy on Val:', accuracy_score(y_test, y_test_pred))
print('Linear Accuracy on Test:', accuracy_score(y_val, y_val_pred), '\n')

y_val_pred = model_poly.predict(x_val)
y_test_pred = model_poly.predict(x_test)
print('Poly Accuracy on Val:', accuracy_score(y_test, y_test_pred))
print('Poly Accuracy on Test:', accuracy_score(y_val, y_val_pred), '\n')

y_val_pred = model_rbf.predict(x_val)
y_test_pred = model_rbf.predict(x_test)
print('RBF Accuracy on Val:', accuracy_score(y_test, y_test_pred))
print('RBF Accuracy on Test:', accuracy_score(y_val, y_val_pred), '\n')
```

### Grid Search
I chose a model that has both parameters C & Gamma -> RBF

```python
C = [0.01, 0.1, 1, 10, 100]
accs = []
for c in C:
    model = SVC(kernel= 'rbf', C=c)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_val)
    acc = accuracy_score(y_val, y_pred)
    accs.append(acc)

plt.plot(C, accs, marker= 'o')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.show()
```

```python
gamma = [0.001, 0.01, 0.1, 1]
accs = []
for g in gamma:
    model = SVC(kernel= 'rbf', gamma= g)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_val)
    acc = accuracy_score(y_val, y_pred)
    accs.append(acc)

plt.plot(gamma, accs, marker= 'o')
plt.xlabel('Gamma')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

params = {'C': C, 'gamma': gamma}

grid = GridSearchCV(SVC(kernel= 'rbf'), params)
grid.fit(x_train, y_train)

best_model = grid.best_estimator_
best_params = grid.best_params_
print(best_params)
```

```python
best_model.fit(x_train, y_train)
pred = best_model.predict(x_test)
accuracy_score(y_test, pred)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```
